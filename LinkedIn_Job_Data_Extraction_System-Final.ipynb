```python
from selenium import webdriver
import time
import pandas as pd
import os
```

```python
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
url1='https://www.linkedin.com/jobs/search?keywords=&location=India&locationId=&geoId=102713980&f_TPR=&f_JT=F%2CC&f_WT=2&position=1&pageNum=0'
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(10)
driver.get(url1)
```

```python
from selenium.webdriver.common.by import By

n = driver.find_elements(By.CLASS_NAME, 'results-context-header__job-count')[0].text
```

```python
from selenium.webdriver.common.by import By

i = 2
while i <= 50: 
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    i += 1
    
    try:
        send = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='See more jobs']"))
        )
        driver.execute_script("arguments[0].click();", send)   
        time.sleep(3)
            
    except Exception as e:
        print("Error or button not found:", e)
        time.sleep(5)
```

```python
companyname=[]
```

```python
n = '20,000+'

n_int = int(n.replace(',', '').replace('+', ''))

companyname = []

from selenium.webdriver.common.by import By

try:
    for i in range(n_int):
        elements = driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')
        if i < len(elements):
            company = elements[i].text
            companyname.append(company)
        else:
            break
except Exception as e:
    print("An error occurred:", e)
```

```python
companyname
```

```python
titlename=[]
```

```python
n_int = int(n.replace(',', '').replace('+', ''))

titlename = []

try:
    for i in range(n_int):
        title_elements = driver.find_elements(By.CLASS_NAME, 'base-search-card__title')
        if i < len(title_elements):
            title = title_elements[i].text
            titlename.append(title)
        else:
            break
except IndexError:
    print("no")
```

```python
titlename
```

```python
job_locations = []

try:
    for i in range(n_int):
        location_elements = driver.find_elements(By.CLASS_NAME, 'job-search-card__location')  # Replace with actual class name
        if i < len(location_elements):
            location = location_elements[i].text
            job_locations.append(location)
        else:
            break
except Exception as e:
    print("Error occurred:", e)
```

```python
job_locations
```

```python
job_post_url = []

try:
    for i in range(n_int):
        link_elements = driver.find_elements(By.CSS_SELECTOR, '.base-card__full-link.absolute.top-0.right-0.bottom-0.left-0.p-0.z-\[2\]')  # Correct CSS selector
        if i < len(link_elements):
            url = link_elements[i].get_attribute('href')
            job_post_url.append(url)
        else:
            break
except Exception as e:
    print("Error occurred:", e)
```

```python
job_post_url
```

```python

companyfinal = pd.DataFrame(companyname, columns=["company"])
titlefinal = pd.DataFrame(titlename, columns=["title"])
locationfinal = pd.DataFrame(job_locations, columns=["location"])
urlfinal = pd.DataFrame(job_post_url, columns=["url"])
```

```python
final_dataframe = pd.concat([companyfinal, titlefinal, locationfinal, urlfinal], axis=1)
```
